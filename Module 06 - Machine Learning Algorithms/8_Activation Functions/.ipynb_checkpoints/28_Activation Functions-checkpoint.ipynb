### Sigmoid/Logistic Activation Function

<img src='a3.png' />

```python
range -> (0,1)

- predict the probability as an output
- binary classification
```

```python
e = 2.72
```

```python
import math

def sigmoid(x):
    return 1/(1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(0.6)
```

```python
sigmoid(-55)
```

### Tanh/Hyperbolic Tangent Function

<img src='s4.png' />

```python
range -> -1 to 1

- all the negative values will be mapped as negative
- all the positive value will be mapped as positive

```

```python
tanh function is used in

- classification(2 classes)
- feed-forward networks
- RNN

```

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x))/(math.exp(x) + math.exp(-x))
```

```python
tanh(-50)
```

```python
tanh(1)
```

```python
tanh(0)
```

### ReLU(Rectified Linear Unit) Activation Function


<img src='a5.png' />

```python
f(x) = max(0,x)

f(-2) => max(0,-2) => 0
f(7)  => max(0,7)  => 7
```

```python
- simple and computationaly efficient
- it will help solving vanishing gradient problem
- faster to give the output in training
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-17)
```

```python
relu(9)
```

```python
relu(100)
```

### Leaky ReLU


<img src='a6.png' />

```python
it will help to increase the range of the function

f(x) = max(a*x, x)

a = 0.1
```

```python
def leaky_relu(x, alpha=0.1):
    return max(x, x*alpha)
```

```python
leaky_relu(100)
```

```python
leaky_relu(-1000)
```

```python
leaky_relu(-17)
```

### Softmax Function


<img src='a7.jpg' />

```python
input -> vector

output -> probability distribution (sum up to 1)
```

```python
- multiclass classification
```

```python
import numpy as np
```

```python
def softmax(x):
    expo = np.exp(x)
    return expo/np.sum(expo, axis=0)
```

```python
x = np.array([1,2,3])

print(softmax(x))
```

```python
sum(softmax(x))
```
